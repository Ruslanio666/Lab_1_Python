### Задание №2:
Написать генератор случайных многомерных матриц, который принимает
опциональные параметры m, n, min_limit, max_limit, где m и n указывают размер
матрицы, а min_lim и max_lim - минимальное и максимальное значение для
генерируемого числа . По умолчанию при отсутствии параметров принимать следующие
значения:
m = 50
n = 50
min_limit = -250
max_limit = 1020

```python
import random

def gArr(m=50, n=50, min_limit=250, max_limit=1020):
    arr = []
    for i in range(m):
        arr.append([0]*n)
    for i in range(m):
        for j in range(n):
            arr[i][j] = random.randint(min_limit, max_limit)
    return arr
```

```python
import timeit
import random

#Сортировка выбором
def sel_sort(b):
    a = b.copy()
    for i in range(len(a) - 1):
        m = i
        j = i + 1
        while j < len(a):
            if a[j] < a[m]:
                m = j
            j = j + 1
        a[i], a[m] = a[m], a[i]
    return a

#Сортировка вставкой
def ins_sort(b):
    a = b.copy()
    for i in range(1, len(a)):
        k = a[i]
        j = i-1
        while j >= 0 and k < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = k
    return a

#Сортировка обменом
def exc_sort(b):
    a = b.copy()
    for i in range(len(a)):
        for j in range(len(a)-i-1):
            if a[j] > a[j+1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a

#Сортировка Шелла
def shell_sort(arr):
    array=arr.copy()
    for i in range(len(array)):
        d=int(len(array[i])/2)
        while d>0:
            for j in range(len(array[i])):
                for h in range(int(j+d),len(array[i]),d):
                    if array[i][j]>array[i][h]:
                        temp=array[i][j]
                        array[i][j]=array[i][h]
                        array[i][h]=temp
            
            d=int(d/2)
    return array


#Турнирная сортировка
def tourn_sort(array):
    arr = array.copy()
    for i in range(len(arr)):
        tourn_subfunc(arr[i])
    return arr


def tourn_subfunc(arr):
    tree = [None] * 2 * (len(arr) + len(arr) % 2)
    index = len(tree) - len(arr) - len(arr) % 2

    for i, v in enumerate(arr):
        tree[index + i] = (i, v)

    for j in range(len(arr)):
        n = len(arr)
        index = len(tree) - len(arr) - len(arr) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1)
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        arr[j] = x
        tree[len(tree) - len(arr) - len(arr) % 2 + index] = None

#Пирамидальная сортировка
def pyr_subfunc(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]

        pyr_subfunc(arr, n, largest)


def pyr_sort(array):
    arr=array.copy()
    for i in range(len(arr)):
        n = len(arr[i])

        for j in range(n, -1, -1):
            pyr_subfunc(arr[i], n, j)

        for j in range(n-1, 0, -1):
            arr[i][j], arr[i][0] = arr[i][0], arr[i][j]
            pyr_subfunc(arr[i], j, 0)
    return arr

#Быстрая сортировка
def quick_sort(a):
    arr = a.copy()
    for i in range(len(arr)):
        quick_subfunc(0, len(arr[i]) - 1, arr, i)
    return arr


def quick_subfunc(fst, lst, array, row):
    first = int(fst)
    last = int(lst)
    middle = int((first + last) / 2)

    while (first < last):

        while (array[row][first] < array[row][middle]):
            first += 1
        while (array[row][last] > array[row][middle]):
            last -= 1
        if (first <= last):
            array[row][first], array[row][last] = array[row][last], array[row][first]
            first += 1
            last -= 1

    if (fst < last):
        quick_subfunc(fst, last, array, row)
    if (first < lst):
        quick_subfunc(first, lst, array, row)

# Проверка
arr = gArr()

startTime = timeit.default_timer()
for i in range(len(arr)):
    sel_sort(arr[i])
print("Сортировка выбором длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
for i in range(len(arr)):
    ins_sort(arr[i])
print("Сортировка вставкой длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
for i in range(len(arr)):
    exc_sort(arr[i])
print("Сортировка обмена длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
shell_sort(arr)
print("Сортировка Шелла длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
tourn_sort(arr)
print("Турнирная сортировка длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
pyr_sort(arr)
print("Пирамидальная сортировка длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
quick_sort(arr)
print("Быстрая сортировка длится " + str(timeit.default_timer() - startTime) + " сек")

mas = arr.copy()
startTime = timeit.default_timer()
mas.sort()
print("Встроенная сортировка длится " + str(timeit.default_timer() - startTime) + " сек")
```
